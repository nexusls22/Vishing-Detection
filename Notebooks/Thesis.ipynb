### The matter of Voice Spoofing and Voice Spoofing Detection

In general voice spoofing refers to copying or impersonating another indivuals voice, in many cases used to deceive others into giving out personal information or validate money transfers. In the year 2020, a CEO was scamed into wiring 220,000€ to a fraudster who used AI-generated voice spoofing to impersonate their boss. Sythetic voices nowadays are advancing with mastering to mimick aspects like accents, intonation, and subtle vocal cues. This social engineering vector is rising which leads to company's realising, that voices are not a ligitimate way to proof your identity anymore.  
    
Voice spoofing, also called voice phishing, is the fastest growing vector in social engineering, and hence a major issue in cyber security and threat protection. According to the Crowdstrike 2025 Global Threat Report vishing attacks grew 442% inbetween the first and second half of 2024, with the first half of 2025 already exceeding all of 2024. Moreover, Cisco Talos reported that 60% of all phishing-related incidents can be accounted to vishing. 
Therefore our goal for the Software Development Project is to detect vishing attacks and distinct between different attack types, measured on audio features of distinct files.

#### Research Papers: 

Recent industry threat intelligence reports provide valuable empirical insights into the evolving cybersecurity landscape. For this study we looked at some of the most influential annual reports, published by IBM X-Force and CrowdStrike, both of which draw on large-scale telemetry and incident response data to identify global attack trends. While both reports aim to characterize modern cyber threats, they differ in analytical focus, methodology, and interpretation of key developments.

The IBM X-Force Threat Intelligence Index emphasizes structural weaknesses in organizational security systems, especially the exploitation of identity systems and its known vulnerabilities. IBM constantly reports, across multiple editions, credential abuse as a dominant vector with reported incidents. Approximately 30% of attacks involving valid account misuse. This concludes that a broader shift happens from exploit-based intrusions toward identity-driven attacks. Hackers nowadays rather try to log in, then to break in. In addition, IBM highlights that fundamental security gaps, such as unpatched systems and misconfigurations persist, noting that vulnerability exploitation accounts for a significant proportion of attacks (over 25-40% depending on the year).

In contrast, the CrowdStrike Global Threat Report greater emphasizes on attacker behavior, speed, and operational sophistication. One of the major findings is the dramatic acceleration of attack timelines, with  breakout times - the window between initial compromise and first lateral movement (navigating deeper into a network after gaining access) - reaching as low as 27 seconds in recent observations. Furthermore, it reports that 82% of detected intrusions are now malware-free, indicating a shift toward hands-on-keyboard techniques and the abuse of legitimate tools. This perspective highlights the increasing importance of real-time detection and response capabilities, rather than traditional signature-based defenses.

Both, IBM and CrowdStrike, reports converge on the growing role of artificial intelligence (AI) in cybercrime, although their interpretations are slightly different. As of IBM, AI primarily is an accelerator of existing attack techniques, enabling faster vulnerability discovery and exploitation.


1. IBM X-Force Threat Intelligence Report: 

Is based on: 
- Incident response cases
- Thread intelligence (dark web, malware, etc.)
- Data from over 100+ countries and billion of daily events.

Goal: 
- Understand how attackers operate
- Identify trends over time
- Provide practical defense recommondations.

2. Core Finding: Identity is #1 Attack Vector
- Abuse of valid login credtials is now the dominant entry method.
- Nearly 30% of attacks involve valid accounts.

This matters because attackers are not longer "breaking in" but they are logging in with valid credentials, blending in with the other legitimate users. 

Trends:
- Nearly 1 in 3 attacks involve stolen credentials.
- Infostealer malware surged massively (up to +266%)

3. CrowdStrike Global Threat report:

- Report Basis: Incident response, threat hunting, and Falcon® platform telemetry tracking over 280 adversaries.

- Goal: Reveal how adversaries exploit trusted tools, AI, and cross-domain access to accelerate attacks.

- Core Finding: AI is now the primary attack accelerant, with 89% surge in AI-enabled attacks and breakout time dropping to 29 minutes (fastest: 27 seconds).

- Key Trends:

    - 82% malware‑free attacks – adversaries rely on valid credentials and legitimate system tools.

    - 37% rise in cloud intrusions; state‑sponsored cloud attacks up 266%.

    - Prompt injection attacks hit over 90 organizations targeting GenAI tools.

    - 42% of vulnerabilities exploited as zero‑days before public disclosure.

    - $1.46 billion stolen in largest single cryptocurrency heist on record.
