## Introduction
<p style="text-align: left;">An expanded version of the abstract minus the conlusions.
Vishing, also called voice phishing, is the fastest growing vector in social engineering, and hence a major issue in cyber security and threat protection. According to the Crowdstrike 2025 Global Threat Report vishing attacks grew 442% inbetween the first and second half of 2024, with the first half of 2025 already exceeding all of 2024. Moreover, Cisco Talos reported that 60% of all phishing-related incidents can be accounted to vishing. 
Therefore our goal for the Software Development Project is to detect vishing attacks and distinct between different attack types, measured on audio features of distinct files.
